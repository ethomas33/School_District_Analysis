```python
import pandas as pd
```

```python
# Files to load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# read schools data in data frame
school_data_df = pd.read_csv(school_data_to_load)
school_data_df
```

```python
#Read the student data file and store in dataframe.
student_data_df = pd.read_csv(student_data_to_load)
student_data_df.head()
```

```python
# Determine if there are any missing values in the school data.
school_data_df.count()
```

```python
# Determine if there are any missing values in the student data.
student_data_df.count()
```

```python
#Determine data types for the student data frame
student_data_df.dtypes
```

```python
# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

# Iterate through the words in the "prefixes_suffixes" list and replace them with an empty space, "".
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"", regex= True)
```

```python
#combine the data into a single dataset
school_data_complete_df = pd.merge(student_data_df, school_data_df, on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
# Get the total number of students
student_count = school_data_complete_df["Student ID"].count()
student_count
```

```python
#calculate the total number of schools
schools_count = school_data_df["school_name"].count()
schools_count
```

```python
#calculate the total number of schools
school_count_2 = school_data_complete_df["school_name"].unique()
school_count_2
```

```python
#length
school_count_2len = len(school_data_complete_df["school_name"].unique())
school_count_2len
```

```python
total_budget = school_data_df["budget"].sum()
total_budget
```

```python
#calculate the average reading score
average_reading_score = school_data_complete_df["reading_score"].mean()
average_reading_score
```

```python
#calculate the average math score
average_math_score = school_data_complete_df["math_score"].mean()
average_math_score
```

```python
#determining passing grade
passing_math = school_data_complete_df[school_data_complete_df["math_score"]>=70]
passing_math
```

```python
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"]>=70]
passing_reading
```

```python
#calculate the number of students passing math

passing_math_count = passing_math["student_name"].count()

#calculate the number of students passing math

passing_reading_count = passing_reading["student_name"].count()
```

```python
print(passing_math_count)
print(passing_reading_count)
```

```python
#calculate the percent that passed math
passing_math_percentage = passing_math_count/float(student_count)*100

#calculate the percent that passed math
passing_reading_percentage = passing_reading_count/float(student_count)*100
```

```python
print(passing_math_percentage)
print(passing_reading_percentage)
```

```python
#calculate the overall passing percentage
passing_math_reading = school_data_complete_df[(school_data_complete_df["math_score"]>=70)&(school_data_complete_df["reading_score"]>=70)]

passing_math_reading.head()
```

```python
#overall count
overall_passing_math_reading_count=passing_math_reading.student_name.count()
overall_passing_math_reading_count
```

```python
# Calculate the overall passing percentage.
overall_passing_percentage = overall_passing_math_reading_count / student_count * 100
overall_passing_percentage
```

```python
# adding a list of values with keys to create a new data frame
district_summary_df = pd.DataFrame(
        [{"Total Schools": schools_count,
         "Total Students": student_count,
         "Average Math Score": average_math_score,
         "Average Reading Score": average_reading_score,
          "% Passing Math": passing_math_percentage,
          "% Passing Reading": passing_reading_percentage,
          "% Overall Passing": overall_passing_percentage}])
district_summary_df
```
